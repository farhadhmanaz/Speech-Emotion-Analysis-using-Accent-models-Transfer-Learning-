```python
import csv
import os
import audioread
import pandas as pd
import numpy as np
import librosa
from tqdm import tqdm
import math
import librosa,librosa.display
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow.keras as keras
import tensorflow as tf

# Getting featues of the audio file
    
def features(file_name):
    audio,sample_rate=librosa.load(file_name,res_type='kaiser_fast')
    mfccs=librosa.feature.mfcc(audio,sr=22050,n_fft=2048,hop_length=512,n_mfcc=13)
    scaled_mfccs=np.mean(mfccs.T,axis=0)
    return scaled_mfccs

    
# prediction for Accent classification

def predict_Accent(model,file,name):
    featuress=[]
    data=features(file)
    featuress.append([data,name])
    data_frame=pd.DataFrame(featuress,columns=['f','c'])
    x_predict=np.array(data_frame['f'].tolist())
    y_predict=np.array(data_frame['c'].tolist())


    x_predict=x_predict[...,np.newaxis]

    prediction=model.predict(x_predict)
    
    index=np.argmax(prediction, axis=1)

    if(index==0):
      val="African_American"
    elif(index==1):
      val="Asian"
    else:
      val="Cauasian"

    return val
```

#Loading Model

```python
new_model = keras.models.load_model("/content/drive/MyDrive/Speech_Emotion_Analysis_17 08 2021/savedModels/savedModelAccent.h5")
```

```python
file="/content/drive/MyDrive/Speech_Emotion_Analysis_17 08 2021/test_files/African_American_angry.wav"
classes=file.split("/")
name=classes[-1]

val=predict_Accent(new_model,file,name)

print("Expected index:  {}".format(name))
print()
print("Predicted index: {}".format(val))
```
