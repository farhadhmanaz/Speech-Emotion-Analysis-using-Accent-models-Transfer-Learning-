```python
import csv
import os
import audioread
import pandas as pd
import numpy as np
import librosa
from tqdm import tqdm
import math
import librosa,librosa.display
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow.keras as keras
import tensorflow as tf
from scipy.io.wavfile import write
```

```python
# Getting featues of the audio file
    
def features(file_name):
    audio,sample_rate=librosa.load(file_name,res_type='kaiser_fast')
    mfccs=librosa.feature.mfcc(audio,sr=22050,n_fft=2048,hop_length=512,n_mfcc=13)
    scaled_mfccs=np.mean(mfccs.T,axis=0)
    return scaled_mfccs

def predict_Emotion(model,file):
    featuress=[]
    data=features(file)
    featuress.append([data])
    data_frame=pd.DataFrame(featuress,columns=['f'])
    x_predict=np.array(data_frame['f'].tolist())
    x_predict=x_predict[...,np.newaxis]
    prediction=model.predict(x_predict)
       
    index=np.argmax(prediction, axis=1)
    if(index==0):
        val="Anger"
    elif(index==1):
        val="Disgust"
    elif(index==2):
        val="Fear"
    elif(index==3):
        val="Happy"
    elif(index==4):
        val="Neutral"
    else:
        val="Sad"

    return val
```
